```python
from PIL import Image
import pytesseract
from deep_translator import GoogleTranslator

# กำหนดเส้นทางไปยัง Tesseract OCR executable ถ้าจำเป็น
# pytesseract.pytesseract.tesseract_cmd = r'path_to_tesseract_executable'
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def ocr_image(image_path):
    # เปิดและอ่านรูปภาพ
    image = Image.open(image_path)
    # ใช้ Tesseract OCR อ่านข้อความจากรูปภาพ (ภาษาไทย)
    text = pytesseract.image_to_string(image, lang='tha')
    return text

def translate_text(text, src_lang='th', dest_lang='en'):
    # แปลข้อความจากภาษาไทยเป็นภาษาอังกฤษ
    translated = GoogleTranslator(source=src_lang, target=dest_lang).translate(text)
    return translated

# กำหนดเส้นทางไปยังรูปภาพ
image_path = r'D:\Documents\AI_hackaton\IMG_Data\10.png'

# อ่านข้อความจากรูปภาพด้วย OCR
thai_text = ocr_image(image_path)
print("ข้อความภาษาไทยที่อ่านได้จากรูปภาพ:")
print(thai_text)

# แปลข้อความจากภาษาไทยเป็นภาษาอังกฤษ
translated_text = translate_text(thai_text)
print("ข้อความที่แปลเป็นภาษาอังกฤษ:")
print(translated_text)
```

```python
import os
from PIL import Image
import pytesseract
from deep_translator import GoogleTranslator
import csv
import re

# กำหนดเส้นทางไปยัง Tesseract OCR executable ถ้าจำเป็น
# pytesseract.pytesseract.tesseract_cmd = r'path_to_tesseract_executable'

def ocr_image(image_path):
    # เปิดและอ่านรูปภาพ
    image = Image.open(image_path)
    # ใช้ Tesseract OCR อ่านข้อความจากรูปภาพ (ภาษาไทย)
    text = pytesseract.image_to_string(image, lang='tha')
    return text

def translate_text(text, src_lang='th', dest_lang='en'):
    # แปลข้อความจากภาษาไทยเป็นภาษาอังกฤษ
    translated = GoogleTranslator(source=src_lang, target=dest_lang).translate(text)
    return translated

def process_images_in_folder(folder_path, output_csv):
    with open(output_csv, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        # เขียน header ลงในไฟล์ CSV
        writer.writerow(['id', 'menu'])
        
        # ดึงชื่อไฟล์ทั้งหมดในโฟลเดอร์และเรียงลำดับตามเลขไฟล์
        filenames = sorted(os.listdir(folder_path))
        
        for filename in filenames:
            # ตรวจสอบว่าชื่อไฟล์มีรูปแบบที่ต้องการหรือไม่
            match = re.match(r"(\d+)\.(png)", filename)
            if match:
                file_number = int(match.group(1))
                if 1 <= file_number <= 137:
                    image_path = os.path.join(folder_path, filename)
                    # อ่านข้อความจากรูปภาพด้วย OCR
                    thai_text = ocr_image(image_path)
                    print(f"{filename}:")
                    print(thai_text)
                    
                    # แปลข้อความจากภาษาไทยเป็นภาษาอังกฤษ
                    translated_text = translate_text(thai_text)
                    print(f"{filename}:")
                    print(translated_text)
                    print("-" * 50)
                    
                    # เขียนข้อมูลลงในไฟล์ CSV
                    writer.writerow([file_number, translated_text])

# กำหนดเส้นทางไปยังโฟลเดอร์ที่มีรูปภาพ
folder_path = 'D:\Documents\AI_hackaton\IMG_Data'
# กำหนดเส้นทางไปยังไฟล์ CSV ที่จะสร้าง
output_csv = 'D:\Documents\AI_hackaton\output1.csv'

# ประมวลผลรูปภาพทั้งหมดในโฟลเดอร์และบันทึกผลลัพธ์ลงในไฟล์ CSV
process_images_in_folder(folder_path, output_csv)
```
